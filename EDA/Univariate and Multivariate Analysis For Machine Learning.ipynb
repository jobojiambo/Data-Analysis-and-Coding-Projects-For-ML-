```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Importing a sample dataset from Scikit Learn
The iris flower is a genus of flowering plants that belong to the family Iridaceae. It is a popular ornamental plant, known for its showy and colorful blooms.The iris flower is also well-known in the field of biology and statistics as it has been used extensively as a model organism for studying genetics and data analysis.

```python
from sklearn import datasets 
```

```python
iris = datasets.load_iris()
# Let's convert to dataframe 
iris = pd.DataFrame(data= np.c_[iris['data'], iris['target']],columns= iris['feature_names'] + ['species'])
iris
```

```python
#Collecting some few inferential statistics
iris.describe()
```

```python
#The columns â€˜speciesâ€™ is categorical, so lets check the frequency distribution for each category.
print(iris['species'].value_counts())
```

```python
# replace the values with class labels 
t = iris.species = np.where(iris.species == 0.0, 'setosa', np.where(iris. species==1.0,'versicolor', 'virginica'))
t
```

```python
# let's remove spaces from column name 
iris.columns = iris.columns.str.replace(' ','') ;iris.describe()
```

```python
j = iris['sepallength(cm)']
k = iris['petallength(cm)']
```

### Pandas dataframe visualization

```python
# Asimple visualization of the sepal length against petal length
plt.scatter(j,k)
```

```python
print(iris['species'].value_counts())
```

```python
# Set the size of the plot 
iris.hist()
plt.suptitle("Histogram", fontsize=16) # use suptitle to add title to all sublots 
plt.show()
```

```python
iris.boxplot()
plt.title("Bar Plot", fontsize=16) 
plt.show()
```

```python
# print the mean length(in cm) for each column by species 
iris.groupby(by = "species").mean()
```

# Multivariate analysis
In multivariate analysis I tried to establish a sense of relationship of all variables with one other

```python
# plot for mean of each feature for each label class 
iris.groupby(by = "species").mean().plot(kind="bar")
plt.title('Class vs Measurements') 
plt.ylabel('mean measurement(cm)') 
plt.xticks(rotation=0)  # manage the xticks rotation plt.grid(True) 
```

```python
# creating correlation matrix 
corr = iris.corr()
corr
```

```python
import statsmodels.api as sm 
sm.graphics.plot_corr(corr, xnames=list(corr.columns)) 
plt.show() 
```

# Pair Plot 
You can understand the relationship attributes by looking at the distribution of the interactions of each pair of attributes. This uses a built-in function to create a matrix of scatter plots of all attributes against all attributes

```python
# use the scatter_matrix function from pandas.plotting
pd.plotting.scatter_matrix(iris, figsize=(10, 10))

# use suptitle to add title to all subplots
plt.suptitle("Pair Plot", fontsize=20)

# display the plot
plt.show()
```
